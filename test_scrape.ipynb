```python
import requests
import tabula.io as tb
import pandas as pd
import pdfplumber
import re
```

```python
url = "https://nuc.edu.ng/wp-content/uploads/2015/09/Sciences%20Draft%20BMAS.pdf"
r = requests.get(url)
with open("Sciences_BMAS.pdf", "wb") as f:
    f.write(r.content)

dfs = tb.read_pdf(
    "Sciences_BMAS.pdf", 
    pages="all", 
    multiple_tables=True,
    encoding="ISO-8859-1",
    silent=True
)
courses = pd.concat(dfs, ignore_index=True)
courses.head()
```

```python
courses.info()
```

```python
clean_df = courses.dropna(axis=1, how='all')
print("Remaining columns:", clean_df.shape)
clean_df.head()
```

```python
clean_df.info()
```

```python
flat_data = []

for _, row in clean_df.iterrows():
    non_nulls = row.dropna().tolist()
    if non_nulls:
        flat_data.append(non_nulls)

flat_df = pd.DataFrame(flat_data)
print(flat_df.shape)
flat_df.head()

```

```python
flat_df.info()
```

```python
flat_df.to_csv("cleaned_bmas_sections.csv", index=False)
```

```python
data = []
with pdfplumber.open("Sciences_BMAS.pdf") as pdf:
    for i in range(32, 34):
        text = pdf.pages[i].extract_text()
        lines = text.split('\n')

        for line in lines:
            match = re.match(r'^([A-Z]{3,4} ?\d{3})\s+(.+?)\s+(\d+)\s+([RCE])\s+(-|\d+)\s+(-|\d+)?$', line.strip())
            if match:
                course_code, title, credit, status, lecture, practical = match.groups()
                data.append({
                    "Course Code": course_code,
                    "Course Title": title,
                    "Credit Unit": int(credit),
                    "Status": status,
                    "Lecture Hours": None if lecture == "-" else int(lecture),
                    "Practical Hours": None if not practical or practical == "-" else int(practical)
                })
df = pd.DataFrame(data)
print(df.head())
df.to_csv("nuc_course_structure.csv", index=False)
```
